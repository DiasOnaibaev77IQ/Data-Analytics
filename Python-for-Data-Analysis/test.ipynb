````python
words = ["apple", "bat", "bar", "atom", "book"]
by_letter = {}
for word in words:
    letter = word[0]  # get the first letter of the word
    if letter not in by_letter:
        by_letter[letter] = [word]  # create a new list for this letter
    else:
        by_letter[letter].append(word)  # add the word to the existing list for this letter
by_letter
````

````python
by_letter.clear()
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)  # setdefault() returns the value of the key if it is in the dictionary; if not, it inserts the key with a default value (in this case, an empty list) and returns that value, then appends the word to the list
by_letter
````

````python
from collections import defaultdict
by_letter = defaultdict(list)  # create a defaultdict where the default value for any key is an empty list
for word in words:
    letter = word[0]
    by_letter[letter].append(word)  # append the word to the list for this letter (if the letter is not already a key in the dictionary, it will be created with an empty list as the default value)
by_letter
````

## Functions

````python
# functions as objects --> in Python, functions are first-class objects, which means they can be passed around and used as arguments just like any other object (string, int, float, list, etc.)
import re
states = ["   Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda", "south   carolina##", "West virginia?"]

def remove_punctuation(value):
    return re.sub("[!#?]", "", value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings(strings, ops):
    result = []
    for value in strings:
        for func in ops:
            value = func(value)
        result.append(value)
    return result

clean_strings(states, clean_ops)
````

### Generators

````python
"""
a generator is a special kind of function that lets you produce a sequence of values over time, 
instead of creating them all at once and saving them in memory.
"""

def ret():
    for i in range(5):
        yield i  # yield is used to create a generator, which is an iterator that generates values on the fly and can be iterated over using a for loop or next() function    

gen = ret()
````

````python
# (value for value in collection) --> generator expression, a concise way to create generators
squares_gen = (x**2 for x in range(1000000))

print(next(squares_gen))  
print(next(squares_gen))  
print(next(squares_gen))  
print(next(squares_gen))  
print(next(squares_gen))  
````

````python
"""Depending on the number of elements produced by the comprehension expression,
the generator version can sometimes be meaningfully faster than the list version,
but the main advantage is that it does not require the memory to store the entire list in memory"""
sum(x**2 for x in range(1000000))  
# or
sum([x**2 for x in range(1000000)]) 
# it will create a list in memory, which can be inefficient for large ranges, 
# while the generator expression will generate each value on the fly and not require additional memory to store the entire list.
````

#### itertools

````python
import itertools
"""
itertools module provides a collection of tools for handling iterators, 
which can be used to create efficient loops and iterators for various tasks, 
such as creating combinations, permutations, and infinite iterators.
"""

def first_letter(word):
    return word[0]

names = ["Alan", "Adam", "Wes", "Will", "Albert", "Steven"]

for letter, names in itertools.groupby(sorted(names), first_letter):
    print(f"{letter}: {list(names)}")
````

## Errors and Exception Handling

````python
try:
    # 1. The "Dangerous" Zone
    # Put code here that might fail (e.g., opening a file, dividing numbers)
    pass

except ValueError:
    # 2. The Contingency Plan
    # Runs ONLY if a 'ValueError' happens inside the try block
    pass

except (TypeError, ZeroDivisionError):
    # You can catch multiple exceptions in a single except block by using a tuple of exception types
    pass

# BETTER: Catching the error object to see Python's built-in message
except ValueError as e:
    print(f"Failed because of: {e}")

else:
    # 3. The "All Clear" Zone
    # Runs ONLY if the try block completed perfectly with ZERO errors
    pass

finally:
    # 4. The "No Matter What" Zone
    # Runs always, whether an error happened or not (used for cleanup)
    pass
````

````python
def set_age(age):
    if age < 0:
        # Stop the program immediately because a negative age makes no sense
        raise ValueError("Age cannot be negative!")
    return age
````

````python
lines = [x.rstrip() for x in open(PATH, encoding='utf-8')]
lines
````

````python
# Closing the file releases its resources back to the operating system, which is important to 
# prevent resource leaks and ensure that other processes can access the file if needed. 
# It also ensures that any changes made to the file are properly saved and that the file is not left in an inconsistent state.
f.close()
````

````python
with open(PATH, encoding='utf-8') as f:
    lines = [x.rstrip() for x in f]
# The with statement automatically takes care of closing the file after the block of code is executed, 
# even if an error occurs within the block.
lines
````

````python
f1 = open(PATH, encoding='utf-8')
f1.read(10)  # read the first 10 characters of the file
# once you have read the first 10 characters, the file pointer is now at the 11th character, 
# so if you call f1.read() again, it will continue reading from where it left off, 
# rather than starting from the beginning of the file.
````

````python
f1.tell()  # returns the current position of the file pointer (in bytes)
````

````python
f2 = open(PATH, mode='rb') # open the file in binary mode (read bytes instead of text)
f2.read(10)  # read the first 10 bytes of the file
````

````python
f2.tell()
````

````python
# file.seek(offset, whence)   
"""
offset --> the number of bytes to move the file pointer
whence --> the reference point for the offset (0 for the beginning of the file, 1 for the current position, 2 for the end of the file)
whence 1, 2 works only on binary files, for text files, only whence 0 is allowed
"""

f1.seek(0)
f1.tell()  # after seeking to the beginning of the file, the file pointer is now at position 0
````

````python
# remove blank lines and write the result to a new file
with open('tmp.txt', mode="w") as handle:
    handle.writelines(x for x in open(PATH) if len(x) > 1)
````
