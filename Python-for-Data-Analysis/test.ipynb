## Dictionary
Stores a collection of key-value pairs, where key and value are Python objects.

```python
dic = {'a': 1, 'b': 2, 'c': 3}
# or
dic = dict(a=1, b=2, c=3)
```

```python
dic['a']  # access value by key
```

```python
'b' in dic  # check if a key is in the dictionary (returns True or False)
```

```python
dic['d'] = 4  # add a new key-value pair to the dictionary
dic
```

```python
del dic['d'] # remove a key-value pair from the dictionary
dic
```

```python
dic.pop('c')  # remove and return the value associated with the key 'c'
```

```python
dic.keys()  # get a view of the dictionary's keys
dic.values()  # get a view of the dictionary's values
dic.items()  # get a view of the dictionary's key-value pairs
```

```python
dic.update({'d': 4, 'e': 5})  # update the dictionary with key-value pairs from another dictionary or iterable of key-value pairs
```

```python
dic
```

```python
dic.update({'d': 6, 'c': 3}) # update the value of an existing key 'd' to 6 and add a new key 'c' with value 3
dic
```

```python
for key, value in dic.items():
    print(f'Key: {key}, Value: {value}')
```

```python
dic.get('f', 'Not found')  # get the value associated with key 'f', or return 'Not found' if 'f' is not in the dictionary
```

```python
words = ["apple", "bat", "bar", "atom", "book"]
by_letter = {}
for word in words:
    letter = word[0]  # get the first letter of the word
    if letter not in by_letter:
        by_letter[letter] = [word]  # create a new list for this letter
    else:
        by_letter[letter].append(word)  # add the word to the existing list for this letter
by_letter
```

```python
by_letter.clear()
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)  # setdefault() returns the value of the key if it is in the dictionary; if not, it inserts the key with a default value (in this case, an empty list) and returns that value, then appends the word to the list
by_letter
```

```python
from collections import defaultdict
by_letter = defaultdict(list)  # create a defaultdict where the default value for any key is an empty list
for word in words:
    letter = word[0]
    by_letter[letter].append(word)  # append the word to the list for this letter (if the letter is not already a key in the dictionary, it will be created with an empty list as the default value)
by_letter
```

## Built-In Sequence Functions

```python
# enumerate() --> adds a counter to an iterable and returns it as an enumerate object, which can be converted to a list of tuples (index, element)
fruits = ["apple", "banana", "cherry"]
print(list(enumerate(fruits)))  # returns a list of tuples (index, element)
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")
```

```python
# sorted --> returns a new sorted list from the items in an iterable
numbers = [5, 2, 9, 1, 5, 6]
sorted(numbers)
```

```python
# zip() --> pairs up the elements of lists, tuples or other sequences
seq1 = ['name', 'surname', 'age']
seq2 = ('Dias', 'Onaibaev', 21)

zipped = list(zip(seq1, seq2))  # returns a list of tuples (index, element)
zipped
```

## Comprehensions

```python
# [expr for value in collection if condition] --> list comprehension, a concise way to create lists
strings = ["a", "as", "bat", "car", "dove", "python"]
tuple([x.upper() for x in strings if len(x) > 2])
```
